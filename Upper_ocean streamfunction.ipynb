```python
# set up the python
import cosima_cookbook as cc
from dask.distributed import Client
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import numpy as np
import xarray as xr
import cmocean as cm
import cartopy.crs as ccrs
import cmocean as cm
import cartopy.feature as cft
import cftime
import IPython.display
import sys, os, warnings
```

```python
client = Client()
client
```

```python
session = cc.database.create_session()
```

```python
# load the MOM6 data
db='/g/data/ik11/outputs/mom6-panan/panant.db' 
expts = ['panant-v2','panant-hycom1','panant-01-zstar']
```

```python
# get the depth
depth = cc.querying.getvar('panant-v2','deptho',session,n=1)
land_mask = np.squeeze(depth.values)*0
land_mask[np.isnan(land_mask)] = 1
land_mask = np.where(land_mask==1,land_mask,np.nan)
```

```python
# time period
start_time = '1996-01-01'
end_time = '1997-01-01'
# set the threshold of density(AABW)
rho_AABW = 1037.125
```

```python
# load ACCESS-OM2 streamfunction psi(x,y,rho,t)
expt = '1deg_jra55v13_iaf_spinup1_B1'
flux1 = cc.querying.getvar(expt,'tx_trans_rho',session,start_time=start_time,end_time=end_time)

# upper ocean flux for the whole time series
flux1 = flux1.sel(potrho = slice(0.,rho_AABW)).sel(time = slice(start_time,end_time))

# annual mean upper ocean streamfunction
flux1_upper = flux1.mean('time').cumsum('grid_yt_ocean').sum('potrho')/1e9

#load the calculation
flux1_upper = flux1_upper.load()
```

```python
# load ACCESS-OM2-025 streamfunction psi(x,y,rho,t)
expt = '025deg_jra55_iaf_omip2_cycle1'
flux2 = cc.querying.getvar(expt,'tx_trans_rho',session,start_time = start_time,end_time=end_time)

# flux of upper ocean for the whole time series
flux2 = flux2.sel(potrho = slice(0,rho_AABW)).sel(time = slice(start_time,end_time))

# calculate the streamfunction of upper ocean
flux2_upper= flux2.mean('time').cumsum('grid_yt_ocean').sum('potrho')/1e9

# load the data
flux2_upper = flux2_upper.load()
```

```python
# load ACCESS-OM2-01 streamfunction psi(x,y,rho,t)
expt = '01deg_jra55v13_iaf'
flux3 = cc.querying.getvar(expt,'tx_trans_rho', session, start_time=start_time, end_time=end_time)

# flux of upper ocean for the whole time series
flux3 = flux3.sel(potrho = slice(0.,rho_AABW)).sel(time = slice(start_time,end_time))

# calculate the streamfunction of upper ocean
flux3_upper = flux3.mean('time').cumsum('grid_yt_ocean').sum('potrho')/1e9

# load the data
flux3_upper=flux3_upper.load()
```

```python
# load MOM6 streamfunction 
flux4 = cc.querying.getvar('panant-v2','umo',session)

flux4 = flux4.sel(rho2_l = slice(0.,rho_AABW)).sel(time = slice(start_time,end_time))

# calculate the psi of upper ocaen
flux4_upper = flux4.mean('time').cumsum('yh').sum('rho2_l')/1e9
flux4_upper = flux4_upper.load()
```

```python
# plot
projection=ccrs.SouthPolarStereo()
fig = plt.figure(figsize=(8, 8), dpi=300)
# 1 deg
ax = plt.subplot(2,2,1,projection = projection)
ax.coastlines(resolution='50m')
ax.add_feature(cft.LAND)
ax.set_extent([-280,80,-82,-50], crs = ccrs.PlateCarree())

theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta),np.cos(theta)]).T
circle = mpath.Path(verts *radius + center)
ax.set_boundary(circle, transform=ax.transAxes)
p1 = flux1_upper.plot.contourf(ax=ax, levels = np.arange(-40,40,2),transform=ccrs.PlateCarree(),
                              cmap=cm.cm.delta,add_colorbar=False)
ax.contourf(depth.xh,depth.yh,land_mask,transform=ccrs.PlateCarree(),colors='grey')
ax.set_title('Streamfunction_upper 1$^{\circ}$')

#0.25 deg
ax = plt.subplot(2,2,2, projection=projection)
ax.coastlines(resolution='50m')
ax.add_feature(cft.LAND)
ax.set_extent([-280, 80, -82, -50], crs=ccrs.PlateCarree())

theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5,0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax.set_boundary(circle, transform = ax.transAxes)
p1 = flux2_upper.plot.contourf(ax=ax, levels=np.arange(-40,40,2),transform = ccrs.PlateCarree(),
                              cmap=cm.cm.delta,add_colorbar = False)
ax.contourf(depth.xh,depth.yh,land_mask,transform = ccrs.PlateCarree(),colors = 'grey')
ax.set_title('Streamfunction_upper 0.25$^{\circ}$')

# 0.1 deg
ax = plt.subplot(2, 2, 3, projection=projection)
ax.coastlines(resolution='50m')
ax.add_feature(cft.LAND)
ax.set_extent([-280, 80, -82, -50], crs=ccrs.PlateCarree())

theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax.set_boundary(circle, transform=ax.transAxes)
p1 = flux3_upper.plot.contourf(ax=ax,levels=np.arange(-40,40,2),transform=ccrs.PlateCarree(),
                      cmap=cm.cm.delta,add_colorbar = False)
ax.contourf(depth.xh,depth.yh,land_mask,transform=ccrs.PlateCarree(),colors='grey')
ax.set_title('Streamfunction_upper 0.1$^{\circ}$')

# 0.05 deg
ax = plt.subplot(2,2,4, projection=projection)
ax.coastlines(resolution='50m')
ax.add_feature(cft.LAND)
ax.set_extent([-280,80,-82,-50],crs=ccrs.PlateCarree())

theta = np.linspace(0,2*np.pi,100)
center,radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
ax.set_boundary(circle, transform=ax.transAxes)
p1 = flux4_upper.plot.contourf(ax=ax, levels = np.arange(-40,40,2),transform=ccrs.PlateCarree(),
                              cmap=cm.cm.delta,add_colorbar = False)
ax.contourf(depth.xh,depth.yh,land_mask,transform=ccrs.PlateCarree(),colors='grey')
ax.set_title('Streamfunction_upper 0.05$^{\circ}$')

# add the colorbar
ax = plt.axes([0.92, 0.37, 0.02, 0.3])
cb = plt.colorbar(p1, cax=ax, orientation='vertical',extend='neither')
cb.ax.set_ylabel('Streamfunction (Sv)')
```
